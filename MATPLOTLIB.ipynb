```python
#object oriented ploting

x =[1,2,3,4,5]
y =[1,2,4,3,6]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlim([0,4])
ax.set_xticks([2,4,6,9])
ax.set_yticks([2,4,6,9])
ax.plot(x,y)
```

```python
#Generating multiple graphs
fig = plt.figure()
fig,ax = plt.subplots(2,3)
ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[0,2].bar(y,x)
ax[1,0].pie(y)
ax[1,1].pie(x)
ax[1,2].plot(x,y)
```

```python
df = pd.read_csv("G:\mtcars.csv")
```

```python
mpg = df.mpg
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,2,1.5])
ax.set_xticks(range(32))
ax.set_xticklabels(df.car_model,rotation = 60,fontsize = 10)
ax.set_title("Miles per Gallon of cars in mtcars",fontsize = 25)
ax.set_xlabel('car names',fontsize=15)
ax.set_ylabel('miles/gal',fontsize = 25)
ax.legend(loc = 'upper right')
ax.plot(mpg)

```
